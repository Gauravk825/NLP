```python
import re

def find_highlighted_word(text):
    # Define the pattern to find highlighted word in <b> tags
    pattern = r'<b>(.*?)</b>'
    
    # Use re.findall to extract all matches
    matches = re.findall(pattern, text)
    
    if matches:
        # Assuming there's only one match, return the first one
        return matches
    else:
        return None
```

3. Use word sense disamguiation algorithm to find the appropriate meaning of the
highlighted word.
1. "He used the key to unlock the door."
"The answer to the problem lies in the key details."
2. The weight is measured on a scale weighing instrument.
The project is too large in scale for one person.
3. "The knife has a very sharp blade."
"His mind was always sharp and quick."

```python
from nltk.wsd import lesk
from nltk.tokenize import word_tokenize

sents = ['He used the key to unlock the door.', "The answer to the problem lies in the key details."]

for sent in sents:
    al = lesk(word_tokenize(sent), 'key')
    print("Key : ",al.definition())

```

```python
sents = ["The weight is measured on a scale weighing instrument.", "The project is too large in scale for one person."]
for sent in sents:
    al = lesk(word_tokenize(sent), 'scale')
    print(al.definition())
```

```python
sents = ["The knife has a very sharp blade.", "His mind was always sharp and quick."]

for sent in sents:
    al = lesk(word_tokenize(sent), 'sharp')
    print(al.definition())
```

4. Use following sentences to find the bag of words using count vectorizer.


    1. The postman delivered the package to the wrong address.
    2. I wrapped a beautiful present for my friend's birthday.
    3. The delivery truck arrived late due to heavy traffic.
    4. We need to check the shipping address before sending the order.
    5. Online shopping offers a wide variety of products with fast

```python
import numpy as np
import pandas as pd
from nltk.tokenize import word_tokenize
```

```python
sent1 = "The postman delivered the package to the wrong address."
sent2 = "I wrapped a beautiful present for my friend's birthday."
sent3 = "The delivery truck arrived late due to heavy traffic."
sent3 = "We need to check the shipping address before sending the order."
sent4 = "Online shopping offers a wide variety of products with fast."
```

```python
from nltk.tokenize import word_tokenize 

# Function for count the  frequency of words
def count_word_frequencies(sentence, columns):
  words = word_tokenize(sentence.lower())
  # word_counts = {}
  # Create a dictionary with keys set to zero
  word_counts = {key: 0 for key in columns}
  # Count word occurrences using a set for efficiency
  for word in words:
    if word in word_counts:
      word_counts[word] += 1  # Increment count if word exists
    else:
      word_counts[word] = 1   # Initialize count to 1 for new words

  return word_counts

```

```python
columns = set(word_tokenize(sent1.lower()) + word_tokenize(sent2.lower()) + word_tokenize(sent3.lower()) + word_tokenize(sent4.lower()) )
columns
```

```python
d1 = count_word_frequencies(sent1,columns)
d2 = count_word_frequencies(sent2,columns)
d3 = count_word_frequencies(sent3,columns)
d4 = count_word_frequencies(sent4,columns)
```

```python
df = pd.DataFrame([d1, d2, d3, d4])
```
